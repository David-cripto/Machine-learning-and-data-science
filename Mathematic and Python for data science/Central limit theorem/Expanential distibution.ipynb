```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

Выберем в качестве примера экспоненциальное распеределение с параметром $\lambda = 1$ и сгенерируем выборку из 1000 значений.

```python
exp_rv = sts.expon()
sample = exp_rv.rvs(1000)
```

Построим гистограмму выборки и теоретическую функцию плотности распределения.

```python
plt.hist(sample, density=True)
x = np.linspace(0,8, 100)
plt.plot(x,exp_rv.pdf(x))
```

Математическое ожидание экспоненциальной случайной величины  $\mathrm{E}[X]=\frac{1}{\lambda}$, дисперсия $\mathrm{Var}[X]=\frac{1}{\lambda^2}$, в данном примере выбрано стандартное распределение с параметром $\lambda=1$, откуда $\mathrm{E}[X]=\mathrm{Var}[X]=1$. Тогда по центральной предельной теореме $\bar{X}_n\approx\sim N(\mathrm{E}[X],\mathrm{Var}[X]/n)$ =$N(1,1/n)$

При $n=5$ $\bar{X}_n=N(1,1/5)$

```python
x = np.linspace(0,3, 100)
```

```python
normal_rv = sts.norm(loc=1, scale=np.sqrt(1/5))
means = [np.mean(exp_rv.rvs(5)) for _ in range(1000)]
plt.hist(means, density=True)
plt.plot(x, normal_rv.pdf(x))
```

При $n=10$ $\bar{X}_n=N(1,1/10)$

```python
normal_rv = sts.norm(loc=1, scale=np.sqrt(1/10))
```

```python
means = [np.mean(exp_rv.rvs(10)) for _ in range(1000)]
plt.hist(means, density=True)
plt.plot(x, normal_rv.pdf(x))
```

При $n=50$ $\bar{X}_n=N(1,1/50)$

```python
normal_rv = sts.norm(loc=1, scale=np.sqrt(1/50))
means = [np.mean(exp_rv.rvs(50)) for _ in range(1000)]
plt.hist(means, density=True)
plt.plot(x, normal_rv.pdf(x))
```

Как видно из графиков, с увеличением $n$ гистограмма всё сильнее приближает теоретический вид функции плотности случайной величины $\bar{X}_n$.
