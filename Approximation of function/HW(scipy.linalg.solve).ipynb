```python
import numpy as np
import math
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
a = np.array([[1, 1],[1, 15]])
b = np.array([f(1), f(15)])
```

```python
from scipy import linalg
sol = linalg.solve(a,b)
sol
```

```python
def g(x):
    return sol[0] + sol[1]*x
```

```python
%matplotlib inline
```

```python
from matplotlib import pylab as plt

arg = np.arange(0, 16, 0.1)
fun = []
for x in arg:
    fun.append(g(x))
fun = np.array(fun)
plt.plot(arg, fun)
plt.show()
```

```python
a = np.array([[1, 1, 1],[1, 8, 64],[1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
```

```python
sol = linalg.solve(a,b)
```

```python
sol
```

```python
def g_1(x):
    return sol[0] + sol[1]*x + sol[2]*x*x
arg = np.arange(0, 16, 0.1)
fun = []
for x in arg:
    fun.append(g_1(x))
fun = np.array(fun)
plt.plot(arg, fun)
plt.show()
```

```python
a = np.array([[1, 1, 1, 1],[1, 4, 16, 64],[1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([f(1), f(4), f(10), f(15)])
```

```python
sol = linalg.solve(a,b)
sol
```

```python
def g_2(x):
    return sol[0] + sol[1]*x + sol[2]*(x**2) + sol[3]*(x**3)
```

```python
arg = np.arange(0, 16, 0.1)
fun = []
for x in arg:
    fun.append(g_2(x))
fun = np.array(fun)
plt.plot(arg, fun)
plt.show()
```
