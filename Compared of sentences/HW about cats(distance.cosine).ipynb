```python
file_obj = open('sentences.txt', 'r')
```

```python
matrix = file_obj.readlines()
```

```python
matrix
```

```python
for line in matrix:
    print(line.strip())
```

```python
new_lines = []
for line in matrix:
    new_lines.append(line.strip())
```

```python
new_lines
```

```python
new_lines2 = []
for x in new_lines:
    new_lines2.append(x.lower())
new_lines = new_lines2
new_lines
```

```python
import re
for x in range(len(new_lines)):
    new_lines[x] = re.split('[^a-z]', new_lines[x])
new_lines
```

```python
new = []
for x in new_lines:
    new_l = []
    for y in x:
        if y == '':
            continue
        new_l.append(y)
    new.append(new_l)
new
```

```python
my_dct = dict()
iterat = 0
for x in new:
    for y in x:
        if (y in my_dct) == True:
            continue
        my_dct[y] = iterat
        iterat += 1
my_dct
```

```python
ar = []
for x in range(len(new)):
    ar.append([0]*len(my_dct))
ar          
```

```python
for y in my_dct:
    for x in range(len(new)):
        iterat = 0
        for z in new[x]:
            if z == y:
                iterat += 1
        ar[x][my_dct[y]] = iterat
ar
```

```python
from scipy.spatial import distance
for x in range(1, len(ar)):
    print(distance.cosine(ar[0],ar[x]))
```
