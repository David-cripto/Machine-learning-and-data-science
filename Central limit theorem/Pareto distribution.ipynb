```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# Распределение Парето
Сгенерируем выборку объёма 1000 из распределения Парето.

```python
b = 2.62
pareto_distr = sts.pareto(b)
sample = pareto_distr.rvs(size = 1000)
```

Гистограмма выборки и теоретическая плотность:

```python
#Гистограмма
plt.hist(sample, density=True, bins=100)
plt.ylabel('Fraction of samples')
plt.xlabel('$x$')

b = 2.62
#Теоретическая плотность
x = np.linspace(sts.pareto.ppf(0.01, b), sts.pareto.ppf(0.99, b), 100)
plt.plot(x, pareto_distr.pdf(x), linewidth=1);
```

Далее генерируем тысячу выборок объема 5 и строим гистограммы их выборочных средних. Так же строим график нормального распределения, приближающего данную выборку:

```python
%%capture
sample_mean = np.empty(1000, dtype=float)

#Генерация тысячи выборок объема 5, находим их среднее
for i in range(1000):
    r = pareto_distr.rvs(size=5)
    sample_mean[i] = np.mean(r)
```

```python
#Рисуем гистограмму
plt.hist(sample_mean, density=True, histtype='stepfilled',bins=100, alpha=0.5)
plt.ylabel('Fraction of samples')
plt.xlabel('$x$')

#Получаем мат.ожидание, дисперсию
mean, var = pareto_distr.stats(moments = 'mv')

#Аппроксимирующее нормальное распределение
norm_dist = sts.norm(loc=mean, scale=math.sqrt(var)/math.sqrt(5))

#Рисуем плотность аппроксимирующего нормального распределения
pdf = norm_dist.pdf(x)
plt.plot(x, pdf, linewidth=2);
```
