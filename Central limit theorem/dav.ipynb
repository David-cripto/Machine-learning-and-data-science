```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

## Равномерное распределение 

```python
a = 1
b = 4
unif = sts.uniform(a, b - a)
# выборка из равномерного распределения
val = unif.rvs(1000)
```

```python
val
```

```python
# график плотности равномерного распределния от 1 до 4
x = np.linspace(0, 5,1000)
pdf = unif.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
# гистограмма
plt.hist(val, normed=True)
```

## Объем выборки n = 10

```python
a = 1
b = 4 
unif = sts.uniform(a, b - a)
# массив выборочных средних
aver = np.array([[]])
# объем выборки из равномерного распределения
n = 10
for un_x in range(1000):
    # выборка из равномерного распределения случайной величины
    val_1 = unif.rvs(size = n)
    # нахожу среднее из выборки с помощью np.mean и добавляю в aver
    aver = np.append(aver, np.mean(val_1))
aver
```

```python
# гистограмма средних
plt.hist(aver, density = True)
```

```python
# EX равномерного распределения
ex = (b + a)/2
# DX равномерного распределения
dx = (b - a)**2 / 12
```

```python
ex 
```

```python
dx
```

```python
# оценка дисперсии выборочных средних
dx_aver = dx/n
dx_aver
```

```python
# среднеквадратичное значение выборочных средних
import math
sig = math.sqrt(dx_aver)
sig
```

```python
mu = ex
sigma = sig

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# график плотности распределния нормально распределенной велечины
x = np.linspace(1.5, 3.7, 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

# гистограмма
plt.hist(aver, density = True)
```
